Problem 1: Even Number Skipper

```python
for i in range(1,21):
  if i % 2 == 0:
    continue
  print(i)
```

Problem 2: Password Checker

```python
correct_password = "DaniAndi"
attempts = 3

while attempts > 0:
      password = input("Enter your password: ")
      if password == correct_password:
          print("Access granted!")
          break
      else:
        attempts -= 1
        print("Access denied. Try again.")
        if attempts == 0:
          print("Maximum attempts reached. Access denied.")
```

Problem 3: Number Filter Display

```python
for i in range(1,10):
    if i == 4:
        continue
    if i == 8:
        break
    if i == 6:
        pass
    print(i)
```
